```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
cat = plt.imread('cat.jpg')
```

```python
plt.imshow(cat)
plt.axis("off")
plt.show()
```

```python
cat = cat.reshape((-1,3))
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=4)
```

```python
km.fit(cat)
```

```python
colors = np.array(km.cluster_centers_,dtype='int32')
print(colors)
print(colors.dtype)
```

```python
i = 1
for color in colors:
    plt.subplot(1,4,i)
    plt.axis("off")
    i+=1
    
    new_img = np.zeros((100,100,3))
    new_img[:,:,:] = color
    plt.imshow(new_img)
plt.show()    
```
